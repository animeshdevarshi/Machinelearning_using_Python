##### Iteration after one hot encoding for age

```python
X = pd.concat([Concrete_data.drop(['Compressive_Strength', 'Age'], axis=1), 
           pd.get_dummies(Concrete_data['Age'], prefix="Age")], axis=1)
y = Concrete_data['Compressive_Strength']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
def r_squared(actual, predicted):
    TSS = np.sum((actual - np.mean(actual))**2)
    print("TSS: ", TSS)
    RSS = np.sum((actual - predicted)**2)
    print("RSS: ", RSS)
    return(1- RSS/TSS)
```

```python
y_pred = reg.predict(X)
```

```python
r_squared(y,y_pred)
```
